# Dataset Description

- Class: Integer variable representing the class or type of wine. It denotes 3 different varieties or origins of wine.
- Alcohol: Quantitative variable measured in percentage representing the alcohol content of the wine.
- MalicAcid: Quantitative variable measured in g/L representing the concentration of malic acid in the wine.
- Ash: Quantitative variable measured in g/L representing the ash content of the wine.
- AlcalinityOfAsh: Quantitative variable representing the alkalinity of ash in the wine, possibly measured in units of pH or some other scale.
- Magnesium: Quantitative variable measured in mg/L representing the concentration of magnesium in the wine.
- TotalPhenols: Quantitative variable measured in mg/L representing the total phenolic content in the wine.
- Flavanoids: Quantitative variable measured in mg/L representing the concentration of flavonoids in the wine.
- NonflavanoidPhenols: Quantitative variable measured in mg/L representing the concentration of non-flavonoid phenols in the wine.
- Proanthocyanins: Quantitative variable measured in mg/L representing the concentration of proanthocyanins in the wine.
- ColorIntensity: Quantitative variable representing the color intensity of the wine, possibly measured on a scale.
- Hue: Quantitative variable representing the hue or tint of the wine, possibly measured in degrees or on a scale.
- OD280/OD315ofDilutedWines: Quantitative variable representing the absorbance ratio of the wine, possibly measured at specific wavelengths.
- Proline: Quantitative variable representing the concentration of proline in the wine, possibly measured in mg/L.

# K-means Clusttering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Extracting features from the dataset
X = wine.drop('Class', axis=1)

# Initializing an empty dictionary to store silhouette scores for different K values
silhouette_scores = {}

# Testing different values of K
for k in range(2, 11):
    # Fitting KMeans clustering model
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    
    labels = kmeans.labels_
    silhouette_scores[k] = silhouette_score(X, labels)

# Finding the K value with the highest silhouette score
best_k = max(silhouette_scores, key=silhouette_scores.get)
best_score = silhouette_scores[best_k]

print(f"Best K value: {best_k}")
print(f"Silhouette Score for best K value: {best_score}")
```

- Optimal K Value: The best K value for the dataset, determined using the silhouette score, is 2. This suggests that the data can be effectively partitioned into two distinct clusters based on the features provided.
- Interpretation of Silhouette Score: The silhouette score of 0.657 for the best K value indicates a reasonable degree of separation between the clusters. A silhouette score close to 1 suggests that the clusters are well-separated, while a score closer to 0 indicates overlapping clusters. In this case, the silhouette score of 0.657 suggests that the clusters are relatively well-separated.

```python
import matplotlib.pyplot as plt

# Extracting K values and corresponding silhouette scores
k_values = list(silhouette_scores.keys())
silhouette_scores_list = list(silhouette_scores.values())

# Plotting K values vs Silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_values, silhouette_scores_list, marker='o', linestyle='-')
plt.title('K Value vs Silhouette Score')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.xticks(np.arange(min(k_values), max(k_values)+1, 1.0))
plt.grid(True)
plt.show()
```

As we can see we got the highest silhouette score for k=2 so we will consider 2 clusters for our K-means algorithm.
- Although there are 3 classes present in the dataset, but for k = 3, silhouette score is less(0.56) so will take 2 clusters only.


```python
# Fitting KMeans clustering model with K=2
kmeans = KMeans(n_clusters=2, random_state=42)
wine['Cluster'] = kmeans.fit_predict(X)

# Visualizing the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Proline', y='ColorIntensity', hue='Cluster', data=wine, palette='viridis')
plt.title('K-means Clustering (K=2)')
plt.xlabel('Proline')
plt.ylabel('ColorIntensity')
plt.show()
```

The clustering is pretty accurate for proline which can be seen above in the plot.

```python
# Fitting KMeans clustering model with K=2
kmeans = KMeans(n_clusters=2, random_state=42)
wine['Cluster'] = kmeans.fit_predict(X)

# Visualizing the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Flavanoids', y='ColorIntensity', hue='Cluster', data=wine, palette='viridis')
plt.title('K-means Clustering (K=2)')
plt.xlabel('Flavanoids')
plt.ylabel('ColorIntensity')
plt.show()
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Initializing an empty dictionary to store silhouette scores for different K values
silhouette_scores_agglo = {}

# Testing different values of K
for k in range(2, 11):
    # Fitting Agglomerative clustering model
    agglo = AgglomerativeClustering(n_clusters=k)
    agglo.fit(X)
    
    labels_agglo = agglo.labels_
    silhouette_scores_agglo[k] = silhouette_score(X, labels_agglo)

# Finding the K value with the highest silhouette score
best_k_agglo = max(silhouette_scores_agglo, key=silhouette_scores_agglo.get)
best_score_agglo = silhouette_scores_agglo[best_k_agglo]

print(f"Best K value for Agglomerative clustering: {best_k_agglo}")
print(f"Silhouette Score for best K value: {best_score_agglo}")
```

- Optimal K Value: The best K value for the dataset, determined using the silhouette score, is 2. This indicates that the data can be effectively clustered into two distinct groups based on the features provided.
- Interpretation of Silhouette Score: The silhouette score of 0.659 for the best K value suggests a reasonable degree of separation between the clusters. This score indicates that the clusters are well-separated and have distinct boundaries.
- Cluster Interpretation: With K=2, the dataset is partitioned into two clusters by the Agglomerative clustering algorithm. Each cluster represents a group of observations with similar characteristics based on the features included in the analysis.
- Comparison with K-means: The silhouette score for Agglomerative clustering (0.659) is slightly higher than that of K-means clustering (0.657). This suggests that Agglomerative clustering may have produced slightly better-defined clusters in this particular dataset.

```python
import matplotlib.pyplot as plt

# Extracting K values and corresponding silhouette scores
k_values_agg = list(silhouette_scores.keys())
silhouette_scores_list_agg = list(silhouette_scores_agglo.values())

# Plotting K values vs Silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_values_agg, silhouette_scores_list_agg, marker='o', linestyle='-')
plt.title('K Value vs Silhouette Score')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.xticks(np.arange(min(k_values), max(k_values)+1, 1.0))
plt.grid(True)
plt.show()
```

As we can see we got the highest silhouette score for k=2 so we will consider 2 clusters for our K-means algorithm.

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

linked = linkage(wine, method='average')
plt.figure(figsize=(10, 6))
dendrogram(linked, p=5, truncate_mode='level')
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

- Distinct Clusters: The dendrogram shows two primary clusters represented by different colors, suggesting a clear division in the data.
- High Joining Distance: The distance at which these primary clusters join is relatively high, indicating a significant dissimilarity between the two groups.
- Sample Distribution: The samples are grouped into two main clusters, aligning with the silhouette score analysis.
- In conclusion, both the dendrogram and the silhouette score suggest that two clusters provide a good fit for the dataset. The dendrogram visually confirms the separation between the two clusters, while the silhouette score quantitatively supports this division. This congruence between the two methods strengthens the conclusion that two clusters are optimal for your dataset.
